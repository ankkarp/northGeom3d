```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

imgL = cv2.imread('../data/mock_test/imgs/leftcamera/Im_L_1.png',0)
imgR = cv2.imread('../data/mock_test/imgs/rightcamera/Im_R_1.png.png',0)

imgL = cv2.resize(imgL,(320,500))
imgR = cv2.resize(imgR,(320,500))

stereo = cv2.StereoBM_create(numDisparities=16, blockSize=15)
disparity = stereo.compute(imgL, imgR)

plt.imshow(disparity,'gray')
plt.show()
```
